```python
URL_base = 'https://www.parabola.unsw.edu.au/'
titles = []
for year in range(1970, 2019):
    volume = year - 1964
    decade_start = year - year % 10
    decade_end = decade_start + 9
    decade_string = str(decade_start) + '-' + str(decade_end)
    
    issue = 1
    issue_exists = True
    while True:
        url = URL_base + decade_string + '/volume-' + str(volume) + '-' + str(year) + '/issue-' + str(issue)
        r = requests.get(url)
        
        if r.status_code != 200:
            break          
        
        soup = BeautifulSoup(r.text, 'lxml')
        articles = soup.find_all('div', attrs={'class': 'views-field views-field-title'})

        for article in articles:
            title = article.find('span').text
            titles.append((title, year, issue))
        
        issue += 1
    print('Scalped titles for {}'.format(year))
print('Finished')
```

```python
relevant_titles = [title for title in titles if ('Competition' not in title[0]) and ('Problems' not in title[0]) and 
                   ('Book Reviews' not in title[0]) and ('Editorial' not in title[0])]
relevant_titles
```
